## 5. Creating Line Charts

##  Sampling Data 
1. Set the seed for the random number generator.
2. Create 32,404,859 random numbers (the length of your ords_prods_merge dataframe) ranging between 0 and 1.
3. Code a logical test comparing each of those numbers to a value of 0.7 (either less than or greater than 0.7).
4. Store the results of the test in a new list, dev, in the form of boolean values (True and False).

```python
# Creating seed
np.random.seed(4)
```

```python
# Creating a list holding True/False value to the test np.random.rant ()<=0.7 
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Checking results in list
dev
```

```python
# Example to see
np.random.rand (10)
```

```python
# Store 70% of the sample in the dataframe big 
big = ords_prods_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small 
small = ords_prods_merge[~dev]
```

```python
# Checking row full dataframe 
len(ords_prods_merge)
```

```python
# Checking rows in both dataframes
len(big) + len(small)
```

```python
# Redusing samples to columns which necessary
df_2 = small[['orders_day_of_week','prices']]
```

```python
df_2.head ()
```

```python
# Create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export line chart
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_price_week.png'))
```
