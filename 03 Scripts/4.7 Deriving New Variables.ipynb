# 2 If-Statements with User-Defined Functions

```python
# Define a function (price lable)
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function (create new column)
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking values in new column
df['price_range'].value_counts(dropna = False)
```

```python
# Checking values in column
df['prices'].max()
```

# 3 If-Statements with the loc() Function

## 3.1. Price label  (for subset)

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Checking values in new column
df['price_range_loc'].value_counts(dropna = False)
```

## 3.2. Price label (for entire dataframe)

```python
df_orders_products_merged.loc[df_orders_products_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_orders_products_merged.loc[(df_orders_products_merged['prices'] <= 15) & (df_orders_products_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

# 4 If-Statements with For-Loops

## 4.1. Busiest  and slowest day of the week

```python
# Frequency of column
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Define a function 
result = []

for value in df_orders_products_merged ["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Check result
result
```

```python
# Create new column with result
df_orders_products_merged["busiest_day"] = result
```

```python
# Check frequency of result
df_orders_products_merged['busiest_day'].value_counts(dropna = False)
```

## 4.2. Two busiest and two slowest days of the week

```python
# Define function
result_2 = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 3 or value == 4:
    result_2.append("Least busy")
  else:
    result_2.append("Regularly busy")
```

```python
# Check result_2
result_2
```

```python
# Create new column with result
df_orders_products_merged["busiest_days"] = result_2
```

```python
# Check frequency of result
df_orders_products_merged['busiest_days'].value_counts(dropna = False)
```

# 4.3. Busiest_period_of_day

```python
# Frequency of column
df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Define a function 
busiest_period_of_day = []

for value in df_orders_products_merged ['order_hour_of_day']:
  if value in [10,11,14,15,13,12,16,9]:
    busiest_period_of_day.append("Most orders")
  elif value in [3,4,2,5,1,0]:
    busiest_period_of_day.append("Fewest orders")
  else:
    busiest_period_of_day.append("Regularly busy")
```

```python
# Cheching result of function 
busiest_period_of_day
```

```python
# Create new column with result 
df_orders_products_merged["busiest_period_of_day"] = busiest_period_of_day
```

```python
# Check frequency of result
df_orders_products_merged['busiest_period_of_day'].value_counts(dropna = False)
```
